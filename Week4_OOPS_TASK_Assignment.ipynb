##### Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed and average_of_vehicle.

```python
class vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

```python
# class object
ve= vehicle("honda", 400, "yellow")
```

##### Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

```python
class car(vehicle):
    def seating_capacity(self, capacity):
        self.capacity = capacity
        return f"{self.name_of_vehicle}, seating capacity is {self.capacity}"
```

```python
c = car("Honda", 300, "Red")
```

```python
c.seating_capacity(6)
```

##### Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

###### Multiple inheritance is basically to inherit multiple classes to a single class

```python
class myclass1:
    def class1(self):
        return "This is first class"

class myclass2:
    def class2(self):
        return "This is my second class"
```

```python
class child_class(myclass1, myclass2):
    def child(self):
        pass
```

```python
ch = child_class()
```

```python
ch.class1()
```

```python
ch.class2()
```

##### Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this class.

###### """getter and setter are basically used to get the value of private methods in a class and to set the values of a private method in a class"""

```python
class myclass:
    
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        
    @property
    def see_details(self):
        return self.__name
    
    @see_details.getter
    def get_details(self):
        return self.__name, self.__age
    @see_details.setter
    def set_details(self, name):
        self.__name = name
```

```python
cls = myclass("Sai", 23)
```

```python
cls.see_details
```

```python
cls.get_details
```

```python
cls.set_details = "mohan"
```

```python
cls.get_details
```

##### Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

###### method overriding is when we override/ over-write the value of the method in the child class aside from default value of the parent class

```python
class class1():
    def hobby(self):
        return "Cricket"
```

```python
class class2(class1):
    def hobby(self):
        return "Movies"
```

```python
student = class2()
```

```python
student.hobby()
```
