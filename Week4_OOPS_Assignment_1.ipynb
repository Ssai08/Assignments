#### Q3. Explain why the __init__() function is used. Give a suitable example.

>The init() function acts as default constructor. Such as when an object is created of the class and called, by default the init() function is called. This helps initialize the class.

```python
class student:
    def __init__(self, name, course, school):
        self.name = name
        self.course = course
        self.school = school
```

```python
std = student()
```

```python
std = student("Sai", "Data Science", "Physics Wallah")
```

> When we called the class for the first time it raised an error saying 3 positional arguments are required. And the second time the object called the class smoothly when the parameters were passed.
Parameters were passed because the init() function called by default when we called the class

#### Q5. What is inheritance? Give an example for each type of inheritance.

> Inheritace is the property by which the child class inherits the property of the parent class.

Single Inheritance

```python
class parent1:
    def parent_1(self):
        return "This is Parent class 1"

class parent2:
    def parent_2(self):
        return "This is Parent class 2"
```

```python
class child(parent1):
    def child1(self):
        return "This is child class"
```

```python
ch = child()
```

```python
ch.parent_1()
```

Multiple Inheritance

```python
class child_class_1(parent1, parent2):
    def child2(self):
        return "This is child class 2."
```

```python
ch1 = child_class_1()
```

```python
ch1.parent_2()
```

```python
ch1.parent_1()
```

```python
ch1.child2()
```

Multi_level Inheritance

```python
class child_class_3(parent1):
    def child3(self):
        pass
```

```python
class child_class_4(child_class_3):
    def child4(self):
        pass
```

```python
ch4 = child_class_4()
```

```python
ch4.parent_1()
```

```python
ch4.child3()
```
