<b>Q2. What happens when an exception is not handled? Explain with an example.

When an exception is not handled, it raises an error and distupts the smooth flow of the programming.<br>For example:

```python
# When we do not handle the exception:
import logging
logging.basicConfig(filename='error.txt', level=logging.DEBUG)
a = 0
b = 2
logging.info('Division')
print(b/a)
```

```python
# When the exception is handled:
try:
    a = 0
    b = 2
    print(b/a)
except ZeroDivisionError as e:
    logging.error(e)
    print(f"The error occured is {e}")
```

<b>Q3. Which Python statements are used to catch and handle exceptions? Explain with an example.

 Generally, 'try', 'except', 'else' and 'finally' are used to catch and handle exception.

try:<br>
    &nbsp;&nbsp;code to be executed<br>
except:<br>
    &nbsp;&nbsp;run if exception arises<br>
else:<br>
    &nbsp;&nbsp;run if there is no exception<br>
finally:<br>
    &nbsp;&nbsp;always run this code<br>

```python
try:
    2/0
except Exception as e:
    logging.error(e)
    print(e)
else:
    logging.info('There is no exception')
    print('There is no exception')
finally:
    logging.info('This is always executed')
    print('This is always executed')
```

<b>Q5. What are custom exceptions in Python? Why do we need custom exceptions? Explain with an example.

Custom exceptions in python are used to raise custom error. In custom exception we can name the exception case and also we can print the error message as we wish.<br>
There are certain cases where we will need custom exception because we might encounter a new situation during our work for which there is no error defined in python.

<b>For example</b> suppose we want to check wheather a student will be able to attend the examination or not. We can use custom classes to solve this situation. Example is shown in next problem.<br>
In this problem we will create an exception class namely verified and a function to showcase how to handle the exception.

<b>Q6. Create a custom exception class. Use this class to handle exceptions.

```python
# For example:
# 
class verified(Exception):
    def __init__(self, message):
        self.message = message

def verify(attendance):
    if attendance<75:
        raise verified('You have not sufficient attendance to attend the examination.')
    elif attendance>100:
        raise verified('Invalid attendace Percentage! Check again.')
    else:
        print('You can sit in the examination. All the best.')
```

```python
try:
    attendance = float(input('Enter your attendance percentage'))
    verify(attendance)
    logging.debug(verify)
except verified as e:
    logging.error(e)
    print(e)
```
