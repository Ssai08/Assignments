#### Q1. What is Abstraction in OOps? Explain with an example.

> Abstraction is used to hide the irrelevant data in order to reduce the complexity. It can be achieved by using abstract class. It is the blueprint of the subclasses to make.

```python
from abc import ABC, abstractmethod
class student(ABC):
    @abstractmethod
    def sname(self):
        pass
    @abstractmethod
    def sid(self):
        pass
    @abstractmethod
    def scourse(self):
        pass
    @abstractmethod
    def sbatch(self):
        pass
```

#### Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.

>Abstraction is a process of hiding unnecessary data and showing only relevant data.

>Encapsulation is binding the data members with member variables. This will avoid the direct access of variables.Encapsulation minimizes your codeâ€™s part revealed to the user.

```python
class car:
    def __init__(self, brand, model, milage, speed):   
        self.__brand = brand
        self.__model = model
        self.__milage = milage
        self.__speed = speed
        
    def see_speed(self):
        return self.__speed
    def modify_speed(self, speed):
        self.__speed = speed if speed > 0 else self.__speed
    def see_milage(self):
        return self.__milage
    def modify_milage(self, milage):
        self.__milage = milage if milage > 0 else self.__milage
    def see_brand(self):
        return self.__brand
    def see_model(self):
        return self.__model
```

```python
c = car("audi", "A8", 25, 200)
```

```python
c.see_speed()
```

```python
c.modify_speed(-45)
```

```python
c.see_speed()
```

```python
c.modify_speed(300)
```

```python
c.see_speed()
```

```python
c.see_milage()
```

```python
c.see_brand()
```

> In the above code we can see the speed of car only using "see_speed" method and modify speed using "modify_speed" method. Here, we encapsulated the class variables such as speed, milage, brand, model etc. and created other functions to access the methods.
